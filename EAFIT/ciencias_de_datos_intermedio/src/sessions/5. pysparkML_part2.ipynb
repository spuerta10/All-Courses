# Databricks

<img src="Databricks_Logo.png" height="450" width ="450">

<br>
<br>

Con orígenes en la academia y la comunidad de código abierto, Databricks fue fundada en 2013 por los creadores originales de Apache Spark™, Delta Lake y MLflow. Como la primera y única plataforma Lakehouse del mundo en la nube, Databricks combina lo mejor de los almacenes de datos y los lagos de datos para ofrecer una plataforma abierta y unificada para datos e IA.

# Azure Databricks

## Cluster:

Es la máquina que se utilizará como computo para los proyectos de Databricks

<br>
<br>

<img src="cluster.jpg" height="600" width ="600">

<br>
<br>

- **Acces Mode:** indica si se permite utilizar el acceso compartido
- **Single User Acces:** Permite que el usuario en un command window pueda acceder a las credenciales de Azure Active Directory de ese usuario, y se pasan automáticamente a Spark, lo que le permite acceder a los datos en Azure Data Lake Storage (ADLS) sin tener que especificar sus credenciales manualmente
- **worker_type:** es el tipo de máquina que tendrá el cluster
consultar la siguiente documentacion para verificar las diferencias entre las familias de máquinas:
https://azure.microsoft.com/en-us/pricing/details/virtual-machines/series/
- **Driver Type:** El Driver Type mantiene la información de estado de todos los Notebooks conectados al clúster. El Driver Type también mantiene SparkContext e interpreta todos los comandos que ejecuta desde un Notebook o una biblioteca en el clúster, y ejecuta el maestro Apache Spark que coordina con los ejecutores de Spark.
- **Runtime:** Los Runtimes de Databricks son el conjunto de componentes principales que se ejecutan en los clusters. Todos los Runtimes de Databricks incluyen Apache Spark y agregan componentes y actualizaciones que mejoran la usabilidad, el rendimiento y la seguridad.
